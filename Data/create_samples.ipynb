```python
import pandas as pd

files = {
    'device_all_uk': 'restricted/examples/device_all_uk_sample.csv',
    'evse_status_gla': 'restricted/examples/evse_status_gla_sample.csv',
    'gla_location_evse': 'restricted/examples/gla_location_evse_join_sample.csv',
    'openstreet_ev': 'restricted/examples/OpenStreetEV_GLA_sample.csv',
}

for name, path in files.items():
    df = pd.read_csv(path)
    print(f'\n{name}:')
    print(list(df.columns))
```

```python
import pandas as pd
import hashlib

def hash_id(value):
    return "ID_" + hashlib.md5(str(value).encode()).hexdigest()[:8].upper()

# ── device_all_uk ──────────────────────────────────────
df1 = pd.read_csv('restricted/examples/device_all_uk_sample.csv')
df1['zapmap_device_uid'] = df1['zapmap_device_uid'].apply(hash_id)
df1.to_csv('restricted/examples/device_all_uk_sample.csv', index=False)
print("Done: device_all_uk")

# ── evse_status_gla ────────────────────────────────────
df2 = pd.read_csv('restricted/examples/evse_status_gla_sample.csv')
df2['uid']      = df2['uid'].apply(hash_id)
df2['evse_uid'] = df2['evse_uid'].apply(hash_id)
df2.to_csv('restricted/examples/evse_status_gla_sample.csv', index=False)
print("Done: evse_status_gla")

# ── gla_location_evse ──────────────────────────────────
df3 = pd.read_csv('restricted/examples/gla_location_evse_join_sample.csv')
df3['location_id'] = df3['location_id'].apply(hash_id)
df3['evse_id']     = df3['evse_id'].apply(hash_id)
df3.to_csv('restricted/examples/gla_location_evse_join_sample.csv', index=False)
print("Done: gla_location_evse")

# ── openstreet_ev ──────────────────────────────────────
df4 = pd.read_csv('restricted/examples/OpenStreetEV_GLA_sample.csv')

# Hash IDs
for col in ['party_id', 'id', 'external_l']:
    df4[col] = df4[col].apply(hash_id)

# Jitter coordinates
import numpy as np
for col in ['coordinate', 'coordina_1']:
    df4[col] = pd.to_numeric(df4[col], errors='coerce')
    df4[col] = (df4[col] + np.random.uniform(-0.005, 0.005, size=len(df4))).round(2)

# Drop sensitive columns
drop_cols = [
    'address', 'city', 'postal_cod',
    'operator_n', 'operator_w', 'operator_l',
    'suboperato', 'subopera_1', 'subopera_2',
    'owner_name', 'owner_webs', 'owner_logo',
    'name', 'directions', 'images'
]
df4.drop(columns=drop_cols, inplace=True)
df4.to_csv('restricted/examples/OpenStreetEV_GLA_sample.csv', index=False)
print("Done: openstreet_ev")

print("\n✅ 所有文件处理完成！")
```
